<h2> Section 2: Exploratory Data Analysis <h2>

To get a broader overview of the data, let's take a look at the average ERA's of starting pitchers and relief pitchers during different decades. Let's also take a look at the standard deviation of these ERA's for the different decades. This will help us to see 1) the central tendency of pitchers and 2) how consistent these pitchers were during this time. 

<h3> 1960's <h3>

```python
df_1960 = df[(df['Year'] >= 1960) & (df['Year'] <= 1969)]

#Starting Pitchers
df_SP_1960 = df_1960[df_1960['Games_Started'] > 0]
mean_1960_SP = df_SP_1960['Earned_Run_Average'].mean()
std_1960_SP = df_SP_1960['Earned_Run_Average'].std()
print("Mean ERA of Starting Pitchers in the 1960s: " + str(mean_1960_SP))
print("Standard Deviation of ERA of Starting Pitchers in the 1960s: " + str(std_1960_SP))

#Relief Pitchers
df_RP_1960 = df_1960[(df_1960['Games_Started'] == 0) & (df_1960['Innings_Pitched'] > 0)]
mean_1960_RP = df_RP_1960['Earned_Run_Average'].mean()
std_1960_RP = df_RP_1960['Earned_Run_Average'].std()
print("Mean ERA of Relief Pitchers in the 1960s: " + str(mean_1960_RP))
print("Standard Deviation of ERA of Relief Pitchers in the 1960s: " + str(std_1960_RP))

#take mean and standard deviation to find ranges
ranges = [[mean_1960_SP - std_1960_SP, mean_1960_SP, mean_1960_SP + std_1960_SP], [mean_1960_RP - std_1960_RP, mean_1960_RP, mean_1960_RP + std_1960_RP]]

#Use longer plotting sequence to have more unique box plot
plt.boxplot(ranges, tick_labels=["Starting Pitchers", "Relief Pitchers"], showmeans=False, vert=False,)
plt.title("ERA Range for SPs and RPs in the 1960's")
plt.xlabel("Earned Run Average")
plt.grid(axis='y')
plt.show()
```

<h3> Section 5: Data Science <h3>

The question I will be analyzing in Section 5 will be different than the questions in Section 4. The question for this secton will be more directly related and relevant to linear regression and prediction. The question I will be answering is: 'Can I predict the amount of wins in a season for a starting pitcher based on their season's earned run average?'. I will use a linear regression model to predict the pitcher's season wins based on their season's ERA. I will also be using starting pitchers who have pitched a full season (over 30 starts). I will plot the real wins and regression line to see our final prediction.

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#filter my dataset to ERAs from 0 to 10 for Full-Season Starting Pitchers 
df_clean = df[(df['Earned_Run_Average'] >= 0) & (df['Earned_Run_Average'] <= 10) & (df['Games_Started'] / df['Games_Played'] >= 0.75) & (df['Games_Started'] >= 30)]

#need to drop all infinite and NAN values
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dropna.html
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.replace.html
df_clean = df_clean.replace([float('inf'), -float('inf')], float('nan')).dropna(subset=['Earned_Run_Average', 'Wins'])

#define independent and dependent variable 
ind = df_clean[['Earned_Run_Average']]
dep = df_clean['Wins']

model = LinearRegression()
model.fit(ind, dep)

#predict wins
prediction = model.predict(ind)

#use longer plotting sequence to create a more detailed plot
plt.figure(figsize=(12, 6))
plt.scatter(ind, dep, alpha=0.7, label='Real Wins')
plt.plot(ind, prediction, color='orange', linewidth=3, label='Regression Line')
plt.xlabel('Earned Run Average (ERA)')
plt.ylabel('Wins per Season')
plt.title('Wins per Season vs. ERA for Starting Pitchers')
plt.legend()
plt.grid(True)
```

As you can see in the figure, the amount of wins per season for a starting pitcher seems to decrease when the pitcher's earned run average increases. This logically makes sense because the more runs you give up per game, the less likely you are to win games. We can use 'model.predict([[]])' to predict the amount of wins for a starting pitcher in a season based on a starting pitcher's earned run average.

<h4> Example: <h4>

```python
model.predict([[1.50], [3.50], [5.00], [7.00]])
```

<h3> Section 6: A Different Data Visualization <h3>

In this section, I want to take a look at Bob Gibson's career wins, strikeouts, and ERA. I want to see how these values changed and rose over his long career.

```python
#sort Bob Gibson's statistics by year: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html
gibson_df = df[df['Name'] == 'Bob Gibson'].sort_values(by='Year')

#cumulative wins will allow me to see growth over time
# https://pandas.pydata.org/docs/reference/api/pandas.core.groupby.DataFrameGroupBy.cumsum.html#pandas.core.groupby.DataFrameGroupBy.cumsum
gibson_df['Cumulative_Wins'] = gibson_df['Wins'].cumsum()

#use longer plotting sequence
plt.figure(figsize=(12, 6))
plt.plot(gibson_df['Year'], gibson_df['Cumulative_Wins'], marker='o', label='Wins', color='blue')
plt.plot(gibson_df['Year'], gibson_df['Strikeouts'], marker='s', label='Strikeouts', color='red')

plt.title("Bob Gibson: Career Wins and Yearly Strikeouts")
plt.xlabel("Year")
plt.ylabel("Count")
plt.grid(True)
plt.legend()
```

In this visualization, I took Gibson's strikeout value per year and overlaid it with his career wins total. This gives us a better idea of how his better years helped his career strikeout total. From the years 1962 - 1970, Gibson had over 200 strikeouts every season, except for 1! Next, we will look at how his yearly ERA changed throughout his career.

```python
#use longer plotting sequence
plt.figure(figsize=(12, 6))
plt.plot(gibson_df['Year'], gibson_df['Earned_Run_Average'], marker='^', linestyle='-', color='green')

plt.title("Bob Gibson: ERA Over Career")
plt.xlabel("Year")
plt.ylabel("ERA")
plt.grid(True)
```

In this visualization, we can see Gibson's yearly ERA throughout his career. In the figure, you can see his 1968 season, where he posted nearly a 1.00 ERA average! Finally, let's look at wins, ERA, and strikeouts all together.

```python
#use longer plotting sequence
plt.figure(figsize=(12, 6))
plt.plot(gibson_df['Year'], gibson_df['Cumulative_Wins'], marker='o', label='Cumulative Wins', color='blue')
plt.plot(gibson_df['Year'], gibson_df['Earned_Run_Average'], marker='^', label='ERA', color='green')
plt.plot(gibson_df['Year'], gibson_df['Strikeouts'], marker='s', label='Strikeouts', color='red')

plt.title("Bob Gibson: Cumulative Wins, ERA, Strikeouts")
plt.xlabel("Year")
plt.ylabel("Stat Value")
plt.grid(True)
plt.legend()
```

In this last visualization, we can get a better idea of how great Bob Gibson was during his career. It also allows viewers to get an idea of how great he was during his best years. He is the St. Louis Cardinal's all-time leader in wins, strikeouts, WAR, and all sorts of statistical categories. Bob Gibson had a career wins total of roughly 250 wins, a strikeout total over 200 in 8 seasons, and an ERA of 1.12 in 1968. These stats prove that Bob Gibson is one of the greatest pitchers of all time. Coming from a Cardinals fan, he is THE greatest pitcher of all time. :)
